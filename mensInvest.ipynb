```python
import pandas as pd

# Substitua 'seuarquivo.csv' pelo nome do arquivo que você enviou
nome_arquivo = 'invest.csv'

# Use o método read_csv para ler o arquivo CSV
data = pd.read_csv('invest.csv', delimiter=';')

# Exiba o DataFrame para verificar se os dados foram carregados corretamente
data.head()
```

```python
# Dicionário com as dicas de investimento
dicas_investimento = {
    'Bitcoin': [
        "Diversifique sua carteira de criptomoedas.",
        "Mantenha-se informado sobre o mercado de criptomoedas.",
        "Use carteiras seguras para armazenar seus Bitcoins."
    ],
    'Nasdaq': [
        "Diversifique em várias ações na Nasdaq.",
        "Avalie empresas com fundamentos sólidos antes de investir.",
        "Acompanhe as notícias econômicas e desenvolvimentos no mercado de ações."
    ],
    'FII': [
        "Diversifique em diferentes tipos de Fundos de Investimento Imobiliário.",
        "Avalie a qualidade dos ativos imobiliários.",
        "Acompanhe os rendimentos gerados pelos seus investimentos."
    ]
}

# Função para obter 3 dicas de investimento com base no tipo de investimento
def obter_dicas_investimento(tipo_investimento):
    return dicas_investimento.get(tipo_investimento, [])

# Iterar pelos dados e atribuir 3 dicas de investimento a cada cliente
for index, row in data.iterrows():
    nome_cliente = row['Cliente']
    tipo_investimento = row['tipoinvest']
    dicas = obter_dicas_investimento(tipo_investimento)

    print(f"Cliente: {nome_cliente} ({tipo_investimento})")
    print("Dicas de Investimento:")
    for i, dica in enumerate(dicas, start=1):
        print(f"{i}. {dica}")
    print("\n")
```
